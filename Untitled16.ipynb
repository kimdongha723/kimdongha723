```python
import pyupbit
```

```python
df = pyupbit.get_ohlcv("KRW-axs", interval="minute15")
df
```

```python
df = df.reset_index()
df['ds'] = df['index']
df['y'] = df['close']
data = df[['ds','y']]
data
```

```python
from fbprophet import Prophet
```

```python
model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=8, freq='H')
forecast = model.predict(future)
```

```python

fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```
